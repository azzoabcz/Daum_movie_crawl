```python
from pandas import Series,DataFrame
import pandas as pd
import numpy as np

import glob
import os
```

```python
file_list = glob.glob('./data/*')
file_list
```

```python
all_df = DataFrame(columns=pd.Index(['movie_id'
                                     ,'title'
                                     ,'title_eng'
                                     ,'grade'
                                     ,'country'
                                     ,'release_date'
                                     ,'rating'
                                     ,'director'
                                     ,'actor'
                                     ,'story'
                                     ,'count_review'
                                    ])
                  )
```

```python
for file in file_list:
    temp_df = pd.read_csv(file)
    all_df = pd.concat([all_df,temp_df]
                       ,sort=False
                      ,ignore_index=True)
```

```python
all_df
```

```python
all_df.to_csv('./concat_data/daum_movie_info.csv'
              ,index=False
              ,encoding='utf-8')
```
